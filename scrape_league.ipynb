This notebook uses op.gg, a 3rd party website, to scrape a sample of high skill and low skill players.

```python
import requests
import math
import os
from selenium import webdriver
import pandas as pd
import time
import glob
```

During the time of scraping there was 1,700,000 people who had a rank in League of Legends. Page x contains the people from rank 1 + 100*(x-1) to 100*(x).  So to scrape the user with ranking x, we need to go to page (1 + floor(x/100))

```python
options = webdriver.ChromeOptions()
options.add_argument('headless')
#start up the webdriver (POINT PATH TO CHROMEDRIVER INSTALLATION)
chromedriver = os.path.join(os.getcwd(), "chromedriver")
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver, chrome_options=options)
```

```python
all_pages = glob.glob('data/pages/*')
```

```python
for page in range(1, 100):
    #check if we already scraped this page
    if ('data/pages/page_{}.csv'.format(str(page)) in all_pages):
        continue
    #get the ranking's corresponding op.gg page
    driver.get("http://na.op.gg/ranking/ladder/page={}".format(str(page)))
    time.sleep(3)
    rankings = []
    usernames = []
    tiers = []
    lps = []
    levels = []
    wins = []
    losses = []
    winrates = []
    for tr in (driver.find_elements_by_tag_name('tr')[1:]):
        cells = tr.find_elements_by_tag_name('td')
        ranking = cells[0].text
        username = cells[1].text
        tier = cells[2].text
        lp = cells[3].text
        level = cells[4].text
        win = cells[5].text.split()[0]
        loss = cells[5].text.split()[1]
        winrate = cells[5].text.split()[2]
        rankings.append(ranking)
        usernames.append(username)
        tiers.append(tier)
        lps.append(lp)
        levels.append(level)
        wins.append(win)
        losses.append(loss)
        winrates.append(winrate)
        #print('{} is ranked {}, level {}, in tier {} with {} lp, has {} wins, {} losses, and {} winrate'.format(username, ranking, level, tier, lp, win, loss, winrate))
    summoners = pd.DataFrame(
    {'username': usernames,
     'ranking': rankings,
     'tier':tiers,
     'lp':lps,
     'level':levels,
     'wins':wins,
     'losses':losses,
     'winrate': winrates
    })
    print('page {} has {} rows'.format(page, summoners.shape[0]))
    summoners.to_csv('pages/page_{}.csv'.format(page))
```
